Shanon Diversity Analysis

```python
from skbio.diversity import alpha_diversity
import plotly.express as px

# Assuming 'ASD_meta' is your DataFrame with abundance data

# Convert abundance columns to numeric type
abundance_columns = ASD_meta.columns.difference(['Taxonomy'])
ASD_meta[abundance_columns] = ASD_meta[abundance_columns].apply(pd.to_numeric, errors='coerce')

# Now proceed with alpha diversity analysis
alpha_diversity_result = alpha_diversity('shannon', ASD_meta.iloc[:, 1:], ids=ASD_meta['Taxonomy'])
alpha_diversity_df = pd.DataFrame(alpha_diversity_result, index=ASD_meta['Taxonomy'], columns=['Shannon_Diversity'])

# Display or save the alpha diversity results
print(alpha_diversity_df)

# Visualize the alpha diversity using Plotly
fig6 = px.bar(alpha_diversity_df, x=alpha_diversity_df.index, y='Shannon_Diversity',
             labels={'Shannon_Diversity': 'Shannon Diversity'},
             title='Alpha Diversity Analysis',
             height=600, color='Shannon_Diversity')
fig6.update_xaxes(tickangle=90)
fig6.show()
```

```python
# Sort the DataFrame by Shannon diversity in descending order and select the top 50 taxa
top_50_alpha_diversity_df = alpha_diversity_df.sort_values(by='Shannon_Diversity', ascending=False).head(100)

# Visualize the top 50 alpha diversity using Plotly (vertical bar plot)
fig7 = px.bar(top_50_alpha_diversity_df, x=top_50_alpha_diversity_df.index, y='Shannon_Diversity',
             labels={'Shannon_Diversity': 'Shannon Diversity'},
             title='Top 100 Taxa - Shannon Diversity',
             height=1000, color='Shannon_Diversity')  # Adjust height as needed
fig7.update_xaxes(tickangle=45)  # Rotate x-axis labels for better readability
fig7.show()
```

```python
# Sort the DataFrame by Shannon diversity in ascending order and select the top 50 taxa
bottom_50_alpha_diversity_df = alpha_diversity_df.sort_values(by='Shannon_Diversity', ascending=True).head(100)

# Visualize the bottom 50 alpha diversity using Plotly (vertical bar plot)
fig8 = px.bar(bottom_50_alpha_diversity_df, x=bottom_50_alpha_diversity_df.index, y='Shannon_Diversity',
             labels={'Shannon_Diversity': 'Shannon Diversity'},
             title='Bottom 100 Taxa - Shannon Diversity',
             height=1000, color='Shannon_Diversity')  # Adjust height as needed
fig8.update_xaxes(tickangle=45)  # Rotate x-axis labels for better readability
fig8.show()
```

Random Forest to classify individuals to group ase on abundance value of bacterial taxa

```python
from sklearn.model_selection import train_test_split
```

```python
data_encoded = pd.get_dummies(ASD_flip_Clean, columns=['group'], drop_first=True)
data_encoded    ##### A=0 and B=1
```

```python
X = data_encoded.drop(['group_B'], axis=1)

# Reset the index to default integer index
X = X.reset_index(drop=True)

y = data_encoded['group_B']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc= RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(X_train, y_train)
```

```python
rfc_prediction= rfc.predict(X_test)
```

```python
#usinf random forest of classifier
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, rfc_prediction))
print(classification_report(y_test, rfc_prediction))
```

Accuracy: The overall accuracy of your model is 0.72, which means it correctly predicted the class for 72% of the instances in the test set.

Precision: For class 0, the precision is 0.60, and for class 1, the precision is 0.88. Precision measures the accuracy of the positive predictions. Higher precision indicates fewer false positives.

Recall (Sensitivity): For class 0, the recall is 0.86, and for class 1, the recall is 0.64. Recall measures the ability of the model to capture all the positive instances. Higher recall indicates fewer false negatives.

F1-Score: The F1-score is the harmonic mean of precision and recall. It provides a balance between precision and recall. For class 0, the F1-score is 0.71, and for class 1, the F1-score is 0.74.
