```python
import plotly.express as px

# Assuming ASD_meta is your DataFrame
corr_matrix = ASD_meta.corr()

# Creating an interactive heatmap
fig = px.imshow(corr_matrix, labels=dict(x="Column", y="Column", color="Correlation"))

# Display the interactive plot

fig.show()
```

```python

import plotly.express as px
N = 50  # You can adjust this value based on your preference
A3 = ASD_meta.sort_values(by='A3', ascending=False).head(N)

# Create an interactive bar chart using Plotly
fig = px.bar(A3, x='Taxonomy', y='A3', text='A3', title='Top {} Bacteria Abundance in A3'.format(N))
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.update_layout(xaxis_title='Bacteria', yaxis_title='Abundance', xaxis_tickangle=-45)
fig.show()

```

```python
individual_columns = ['A3', 'A5', 'A6', 'A9', 'A31', 'A51', 'A52', 'A53', 'A54',
                       'A59', 'A67', 'A68', 'A69', 'A71', 'A73', 'A76', 'A78', 'A87', 'A89',
                       'A93', 'A101', 'A109', 'A113', 'A114', 'A115', 'A142', 'A144', 'A149',
                       'A164', 'A165', 'B1', 'B2', 'B3', 'B5', 'B6', 'B7', 'B8', 'B13', 'B14',
                       'B28', 'B29', 'B36', 'B37', 'B94', 'B99', 'B103', 'B106', 'B111',
                       'B114', 'B115', 'B120', 'B127', 'B132', 'B141', 'B142', 'B143', 'B152',
                       'B156', 'B158', 'B164']

# Summing the abundance across all individuals for each bacterium
ASD_meta['TotalAbundance'] = ASD_meta[individual_columns].sum(axis=1)

# Selecting the top 20 most abundant bacteria
df_top_50 = ASD_meta.nlargest(50, 'TotalAbundance')

# Create an interactive bar chart using Plotly
fig = px.bar(df_top_50, x='Taxonomy', y='TotalAbundance', text='TotalAbundance',
             title='Top 20 Most Abundant Bacteria Across All Individuals')
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.update_layout(xaxis_title='Bacteria', yaxis_title='Total Abundance', xaxis_tickangle=-45)
fig.show()
```
